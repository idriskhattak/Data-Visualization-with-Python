```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=sns.load_dataset('flights')
```

```python
#lets do a line plot
plt.plot(df.passengers)
```

you can see the trend line by that we can say passengers are increasing through passing year

```python
#now lets do heatmap
flight=sns.load_dataset('flights').pivot('month','year','passengers')#pivot is used to make the data in metirx type of form as you can seen below
```

```python
flight
```

```python
#For usinf heatmap you have to make your data into matrix form as done in the above cell
plt.title('Number of passenger in 1000,s')
sns.heatmap(flight)
```

now you can see more clearly the number of passengers per year

```python
plt.title('Number of passenger in 1000,s')
sns.heatmap(flight, fmt='d', annot=True,cmap='Blues');
```

we can use annot argument to see values inside heatmap and cmap to change the colour of heatmap
