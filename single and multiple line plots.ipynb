```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Single Line Ploting

```python
apple=[1,2,3,4,6,9]
year=[2001,2002,2003,2004,2005,2006]
plt.plot(year,apple);
```

```python
plt.xlabel('Year')
plt.ylabel('Number of apples per ton')
plt.plot(year,apple);
```

Multiple line ploting

```python
years=range(2000,2010)
apples=[2,2,4,3,5,6,5,4,8,9]
oranges=[1,3,5,4,7,9,6,10,12,13]
plt.xlabel('Years')
plt.ylabel("Fruits(Per Ton)")
plt.plot(years,apples,marker='o')
plt.plot(years,oranges,marker='x')
plt.title('Fruit Analysis')
plt.legend(['Apples',"Oranges"])
```

We can use seaborn alos for styling here is an example

```python
sns.set_style('darkgrid')
years=range(2000,2010)
apples=[2,2,4,3,5,6,5,4,8,9]
oranges=[1,3,5,4,7,9,6,10,12,13]
plt.xlabel('Years')
plt.ylabel("Fruits(Per Ton)")
plt.plot(years,apples,marker='o')
plt.plot(years,oranges,marker='x')
plt.title('Fruit Analysis')
plt.legend(['Apples',"Oranges"])
```
