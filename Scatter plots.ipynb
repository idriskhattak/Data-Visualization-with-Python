```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
flower_df=sns.load_dataset('iris')
flower_df.to_csv('flowers')
```

```python
flower_df.head()
```

```python
plt.plot(flower_df.sepal_length,flower_df.sepal_width)
```

```python
#now using scatter plot and passing the same variable
sns.scatterplot(x=flower_df.sepal_length,y=flower_df.sepal_width);
```

```python
#we can also colour the plot using hue funtion and we can also increase the sixe of the dot with s argument
plt.figure(figsize=(12,6))
plt.title("sepal flower dimention")
sns.scatterplot(x=flower_df.sepal_length,y=flower_df.sepal_width,
               hue=flower_df.species,
                s=100);
```
